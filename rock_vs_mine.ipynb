```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  accuracy_score
```

```python
df =pd.read_csv('sonar.csv')
```

```python
#seperating the data 
x =df.drop(columns=['R'],axis=1)
y=df['R']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
model =LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_train_pred = model.predict(x_train)
acc_training_data = accuracy_score(y_train,y_train_pred)
print("trianing_data_acc----->",acc_training_data)

y_test_pred = model.predict(x_test)
acc_test_data = accuracy_score(y_test,y_test_pred)
print("test_data_acc----->",acc_test_data)
```
